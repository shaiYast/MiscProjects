```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/bellemployeesalaries.csv')

# Filter for city 'Bell'
bell_salaries_df = df[df['cityName'] == 'Bell']

print("Columns in the filtered dataset (City = 'Bell'):")
print(bell_salaries_df.columns.tolist())

# Display the first few rows to understand the data
print("\nFirst 5 rows of the filtered dataset:")
print(bell_salaries_df.head())
```

```python
min_wage = bell_salaries_df['wages'].min()
max_wage = bell_salaries_df['wages'].max()

# Create bins with a size of $50,000
bins = np.arange(0, max_wage + 50000, 50000)

plt.figure(figsize=(12, 7))
plt.hist(bell_salaries_df['wages'], bins=bins, edgecolor='black', alpha=0.7)
plt.title('Distribution of Wages for Employees in Bell (Bin Size = $50,000)')
plt.xlabel('Wages')
plt.ylabel('Number of Employees')
plt.xticks(bins, rotation=45, ha='right') # Set x-ticks to bin edges for clarity
plt.grid(axis='y', alpha=0.75)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
BellWages = bell_salaries_df['wages']
print(BellWages.mean())
print(BellWages.median())
print(BellWages.std())
print(BellWages.max())
```

```python
duarte_salaries_df = df[df['cityName'] == 'Duarte']
DuarteWages = duarte_salaries_df['wages']
print(DuarteWages.mean())
print(DuarteWages.median())
print(DuarteWages.std())
print(DuarteWages.max())
```

```python
print("Bell: " + str(((BellWages.max()-BellWages)/BellWages.std()).mean()))
print("Duarte: " + str(((DuarteWages.max()-DuarteWages)/DuarteWages.std()).mean()))
```
